# Support Vector Regressor

```python
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVR

dataset_x= df_forest_with_redlist[['Year','Country','Value']];
dataset_y=df_forest_with_redlist['redlist']

x_train, x_test, y_train, y_test = train_test_split(dataset_x, dataset_y, test_size = 0.2)


#model
reg = make_pipeline(StandardScaler(), SVR(C=1.0, epsilon=0.1))
reg.fit(x_train, y_train)
y_pred=reg.predict(x_test)


print("MAe", metrics.mean_absolute_error(y_test, y_pred))
print("Mse", metrics.mean_squared_error(y_test, y_pred))
print("RMSE", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

#R-square value
print("R squared value:", metrics.r2_score(y_test, y_pred))
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

dataset_x= df_forest_with_redlist[['Year','Country','Value']];
dataset_y=df_forest_with_redlist['redlist']

x_train, x_test, y_train, y_test = train_test_split(dataset_x, dataset_y, test_size = 0.2)

#model
randomf=  RandomForestRegressor(n_estimators = 1000, random_state=42)
randomf.fit(x_train, y_train)
y_pred= randomf.predict(x_test)

print("MAe", metrics.mean_absolute_error(y_test, y_pred))
print("Mse", metrics.mean_squared_error(y_test, y_pred))
print("RMSE", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

#R-square value
print("R squared value:", metrics.r2_score(y_test, y_pred))
```

# Adaboost Regressor

```python
from sklearn.ensemble import AdaBoostRegressor

dataset_x= df_forest_with_redlist[['Year','Country','Value']];
dataset_y=df_forest_with_redlist['redlist']

x_train, x_test, y_train, y_test = train_test_split(dataset_x, dataset_y, test_size = 0.2)

#model
adab = AdaBoostRegressor(learning_rate=0.001, random_state=42, n_estimators=500)
adab.fit(x_train, y_train)
y_pred= adab.predict(x_test)

print("MAe", metrics.mean_absolute_error(y_test, y_pred))
print("Mse", metrics.mean_squared_error(y_test, y_pred))
print("RMSE", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

#R-square value
print("R squared value:", metrics.r2_score(y_test, y_pred))
```
